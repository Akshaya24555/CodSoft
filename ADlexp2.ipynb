```python
#24555A4209

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

STEP 2: LOAD THE DATASET

```python
df = pd.read_csv('housing_data.csv')
df.shape
```

```python
 #Dataset Info
df.info()
```

```python
 #Head of the Dataset
df.tail()
```

STEP 3: APPLY PRE-PROCESSING

Drop any Null Values

```python

df = df.dropna()
df.info()
```

```python
df = pd.get_dummies(df, columns=['24'], drop_first=True).astype(int)
df.tail()
```

Split X features & y target

Applying One Hot Encoding to the Categorical Columns

```python
# Corrected: If '396.9' is the target, it should not be in X.
X_df = df.drop(columns=['396.9', '4.98'])
y_df = df['396.9']
```

```python
X_df.shape, y_df.shape
```

Scale the Featues using MinMax Saling

```python
# Import MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# Scaling X data
scaler_X = MinMaxScaler()
X = scaler_X.fit_transform(X_df)
X[0:1]
```

```python
# Scaling y data
scaler_y = MinMaxScaler()
# 'reshape(-1, 1)' converts the 1-D Series into a 2-D array, since scalers require 2-dimensional input.
y = scaler_y.fit_transform(y_df.values.reshape(-1, 1))
y[0:8]
```

STEP 4: SPLIT DATA INTO TRAINING & TESTING DATASETS

```python
# Import Train Test Split
from sklearn.model_selection import train_test_split
```

```python
# Split Data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=101)
```

```python
print(f"Train shape: {X_train.shape},{y_train.shape}, Test shape: {X_test.shape}, {y_test.shape}")
```

STEP 5: BUILD & TRAIN THE MODEL

```python
# import the model
from sklearn.neural_network import MLPRegressor
```

```python
# Build the Model
model = MLPRegressor(hidden_layer_sizes=(128,64,32),
                     activation='relu',
                     learning_rate_init=0.002,
                     max_iter=800,
                     solver='adam',alpha=0.0002)
```

Train the model

```python
# Use .ravel() to convert y_train from a column vector (n_samples, 1) to a 1D array (n_samples,)
model.fit(X_train,y_train.ravel())
```

STEP 6: EVALUATE THE MODEL

```python
# Generate Predictions
y_pred = model.predict(X_test)
y_pred[0:3]
```

```python
y_pred = scaler_y.inverse_transform(y_pred.reshape(-1,1))
y_test = scaler_y.inverse_transform(y_test.reshape(-1,1))
```

```python
y_pred[0:3], y_test[0:3]
```

```python
# Import the Regression Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Mean Absolute Error
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
# Mean Square Error
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
# RMSE
rmse = np.sqrt(mse)
rmse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

STEP 7: PREDICT A NEW HOUSE PRICE

```python
# STEP 7: PREDICT A NEW HOUSE PRICE (FIXED)

import numpy as np

# Define raw input for a new house based on original columns before one-hot encoding
# The value for column '24' (original MEDV) is assumed to be 21.6 for this example.
# '396.9' (B) is the target, so it's not included here as a feature.
# '4.98' (LSTAT) was dropped from X, so it's also not included here as a feature.
new_house_data_raw = pd.DataFrame([{
    '0.00632': 0.00632,   # CRIM
    '18': 18,            # ZN
    '2.31': 2.31,        # INDUS
    '0': 0,              # CHAS
    '0.538': 0.538,      # NOX
    '6.575': 6.575,      # RM
    '65.2': 65.2,        # AGE
    '4.09': 4.09,        # DIS
    '1': 1,              # RAD
    '296': 296,          # TAX
    '15.3': 15.3,        # PTRATIO
    # '396.9': 396.9,    # This is the target 'y', so not part of X features.
    # '4.98': 4.98,      # This column was dropped from X.
    '24': 21.6           # Original MEDV value, will be one-hot encoded
}])

# Get the columns from the original X_df to ensure new_house_df matches
X_columns = X_df.columns

# Apply one-hot encoding to the '24' column in new_house_data_raw
# It's important to use the same parameters as used for the training data (df)
new_house_processed = pd.get_dummies(new_house_data_raw, columns=['24'], drop_first=True).astype(int)

# Align columns of new_house_processed with X_columns
# This fills any missing columns with 0 and keeps the order consistent
new_house = new_house_processed.reindex(columns=X_columns, fill_value=0)

# Display the prepared new_house DataFrame
print(new_house.shape)
new_house.head()
```

```python
# Scale new input
new_house_scaled = scaler_X.transform(new_house)
```

```python
# Make prediction
predicted_price_scaled = model.predict(new_house_scaled)

# Inverse transform the prediction to get the actual price
predicted_price = scaler_y.inverse_transform(predicted_price_scaled.reshape(-1, 1))

print(f"Predicted house price: ${predicted_price[0][0]:.2f}")
```

```python
# Predict
pred_scaled = model.predict(new_house_scaled)
```

```python
# Inverse scale
pred_original = scaler_y.inverse_transform(pred_scaled.reshape(-1,1))
pred_original
```
