```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# Read CSV file
df = pd.read_csv("session_1.csv")
```

```python
# Create Status column
df["Status"] = df["Duration (minutes).1"].apply(
    lambda x: "Present" if x >= 90 else "Absent"
)
```

```python
# Features and Target
X = df[["Duration (minutes).1"]]
y = df["Status"].map({"Present": 1, "Absent": 0})
```

```python
# Decision Tree Model
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)
```

```python
# Prediction
df["Prediction"] = model.predict(X)
df["Prediction"] = df["Prediction"].map({1: "Present", 0: "Absent"})
```

```python
# Display Student Status
for index, row in df.iterrows():
    print(row["Name (original name)"], ":", row["Prediction"])
```

```python
# Present Students
print("\nPresent Students:")
print(df[df["Prediction"] == "Present"]["Name (original name)"])
```

```python
# Absent Students
print("\nAbsent Students:")
print(df[df["Prediction"] == "Absent"]["Name (original name)"])
```

```python
# Eligible and Not Eligible Students
eligible = sum(df["Duration (minutes).1"] >= 90)
not_eligible = sum(df["Duration (minutes).1"] < 90)
```

```python
print("\nEligible Students :", eligible)
print("Not Eligible Students :", not_eligible)
```
